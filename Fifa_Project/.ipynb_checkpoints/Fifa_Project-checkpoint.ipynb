```python
class listener(StreamListener):
 
    def on_data(self, data):
        try:
            
            jsonData = json.loads(data)
            #print(jsonData)
            tweet = jsonData['text']
            lang = jsonData['lang']
            coordinates = str(jsonData["coordinates"])
            twitter_data = str(time.time()) + " :: " + tweet + " :: " + coordinates + " :: " + lang
            twitter_data.encode("utf-8")
            print(twitter_data)
            
            f = open("worldcup.json","a")
            f.write(str(twitter_data.encode('ascii','ignore')))
            f.write("\n")
            f.close()
            return True
        
        except BaseException as e:
            print("Error on_data: %s" % str(e))
            time.sleep(5)
 
    def on_error(self, status):
        print(status)
```

```python
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
twitterStream = Stream(auth, listener())
twitterStream.filter(track=["FIFAFanFest","WM2018","Russia2018WorldCup","World Cup 2018 Russia"])
```
